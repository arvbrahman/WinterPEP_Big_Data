Analysis and prediction of heart disease on the basis of chest pain and heart beats per minute using machine learning classification algorithms

#Importing Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.model_selection
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
```

#Load the Dataset

```python
df = pd.read_csv('heart.csv')
df.head()
```

# Separating Features and Targets

```python
X = df[['age','cp','thalach']]    #Features
y = df['target']     #Target
X,y
```

# Selection of algorithm and model fitting

```python
model = LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
model.fit(X_train, y_train)
```

```python
ytrain_pred = model.predict(X_train)
ytest_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:5]
y_proba
```

#Accuracy for training data and testing data

```python
print(f"Accuracy for training data: {accuracy_score(y_train,ytrain_pred):.2f}")
print(f"Accuracy for testing data: {accuracy_score(y_test,ytest_pred):.2f}")
```

#Evaluation of model

```python
print(f"Confusion Matrix for training data: \n{confusion_matrix(y_train,ytrain_pred)}")
print(f"Confusion Matrix for testing data: \n{confusion_matrix(y_test,ytest_pred)}")
```

```python
model.score(X_train,y_train)
```

# Taking user inputs

```python
def predict_heart_disease():
    age = int(input("Enter your age: "))
    cp = int(input("Enter your chest pain type (0-3): "))
    thalach = int(input("Enter your maximum heart rate achieved: "))

    user_data = np.array([[age, cp, thalach]])
    prediction = model.predict(user_data)

    result = "You have heart disease." if prediction[0] == 1 else "You don't have heart disease."
    print(result)


predict_heart_disease()
```

#Model Saving

```python
import joblib
joblib.dump(model,'heart_disease_model.pkl')
```

# Model Deployment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.model_selection
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('heart.csv')
df.head()
X = df[['age','cp','thalach']]    #Features
y = df['target']     #Target
model = LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
model.fit(X_train, y_train)
ytrain_pred = model.predict(X_train)
ytest_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)[:5]
y_proba
def predict_heart_disease():
    age = int(input("Enter your age: "))
    cp = int(input("Enter your chest pain type (0-3): "))
    thalach = int(input("Enter your maximum heart rate achieved: "))

    user_data = np.array([[age, cp, thalach]])
    prediction = model.predict(user_data)

    result = "You have heart disease." if prediction[0] == 1 else "You don't have heart disease."
    print(result)

predict_heart_disease()
import joblib
joblib.dump(model,'heart_disease_model.pkl')
```
